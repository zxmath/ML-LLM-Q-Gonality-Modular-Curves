```python
import pandas as pd
import sys
import os
sys.path.append(os.path.join(os.getcwd(), '..', 'src'))
from utils.data_process import data_process
# change the data path to the path of the data
q_gonality_same, q_gonality_diff = data_process(data_path='/home/zxmath/ML-LLM-Q-Gonality-Modular-Curves/data/combined_data_7.h5')

q_gonality_same.head()

# note we don't drop duplicate yet
```

```python
# plot graph of psl2index vs q_gonality

import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.scatter(q_gonality_same['psl2index'], q_gonality_same['q_gonality'], alpha=0.5)
plt.xlabel('psl2index')
plt.ylabel('q_gonality')
plt.title('psl2index vs q_gonality')
plt.show()
```

```python
q_gonality_same
```

```python
q_gonality_same.drop_duplicates(subset=['level','coarse_level','coarse_class_num',
'rational_cusps','rank','log_conductor','cusps','genus','q_gonality'], inplace=True)   

q_gonality_diff.drop_duplicates(subset=['level','coarse_level','coarse_class_num',
'rational_cusps','rank','log_conductor','cusps','genus'], inplace=True)   
print(q_gonality_same.shape)
print(q_gonality_diff.shape)





```

```python
q_gonality_same['level'].value_counts()
# draw graph of level distribution

import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.hist(q_gonality_same['level'], bins=100)
plt.xlabel('level')
plt.ylabel('count')
plt.title('level distribution')
plt.show()

plt.figure(figsize=(10, 6))
plt.hist(q_gonality_diff['level'], bins=100)
plt.xlabel('level')
plt.ylabel('count')
plt.title('level distribution')
plt.show()
```

```python
# draw graph of q_gonality of every 100  
for i in range(200, 1600, 200):
    plt.figure(figsize=(10, 6))
    plt.hist(q_gonality_same[(q_gonality_same['q_gonality']<i) & (q_gonality_same['q_gonality']>i-200)]['q_gonality'], bins=100)
    plt.xlabel('q_gonality')
    plt.ylabel('count')
    plt.title(f'q_gonality distribution when q_gonality < {i} and q_gonality > {i-100}')
    plt.show()



```

```python
print(len(q_gonality_same[q_gonality_same['level']>59])/len(q_gonality_same))

level_small = q_gonality_same[q_gonality_same['level']<=59]
level_large = q_gonality_same[q_gonality_same['level']>59]







```

```python
# draw level_small q_gonality distribution

import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.hist(level_small['q_gonality'], bins=100)
plt.xlabel('q_gonality')
plt.ylabel('count')
plt.title('level <= 59 q_gonality distribution')
plt.show()
```

```python
# draw level_big q_gonality distribution

plt.figure(figsize=(10, 6))
plt.hist(level_large['q_gonality'], bins=100)
plt.xlabel('q_gonality')
plt.ylabel('count')
plt.title('level > 59 q_gonality distribution')
plt.show()
```
